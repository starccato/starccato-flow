```python
# Generate and save epoch plots at lower resolution for smoother GIF
epoch_dir = "../outdir/flow_matching/gif_data"
n_samples = 18000
num_epochs = 256
frame_skip = 4  # Only save every 4th epoch to reduce GIF size

print(f"Generating {num_epochs // frame_skip} epoch frames to {epoch_dir}...")
for epoch in range(num_epochs):
    # Skip epochs to reduce total frames
    if epoch % frame_skip != 0:
        continue
        
    supernovae_complex.sample_supernovae_for_epoch(
        epoch=epoch,
        n_samples=n_samples,
        num_epochs=num_epochs,
        exponential=True,
        epoch_dir=epoch_dir,
        font_family=font_family,
        font_name=font_name
    )

# Now compile saved plots into GIF with optimization
print("Compiling epoch plots into GIF...")
supernovae_complex.create_epoch_gif(
    epoch_dir=epoch_dir,
    output_fname="../../poster-presentation/supernovae_curriculum.gif",
    duration=200,  # Slower playback (200ms per frame)
    num_epochs=num_epochs
)
```
