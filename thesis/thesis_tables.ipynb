# CVAE Architecture Table
Generate a publication-quality LaTeX table of the CVAE model architecture

```python
from starccato_flow.nn.cvae import ConditionalVAE
from starccato_flow.utils.defaults import Y_LENGTH, HIDDEN_DIM, Z_DIM

# Initialize CVAE with your configuration
y_length = Y_LENGTH
hidden_dim = HIDDEN_DIM
z_dim = Z_DIM
param_dim = 4  # Adjust based on your parameters (e.g., 4 for beta, omega, A, Ye)

model = ConditionalVAE(
    z_dim=z_dim,
    hidden_dim=hidden_dim,
    y_length=y_length,
    param_dim=param_dim
)

# Generate LaTeX table
latex_table = model.get_architecture_table()
print(latex_table)
```

## Save LaTeX to File
Optionally save the LaTeX table to a file for inclusion in your thesis

```python
# Save to file
output_path = "../../2025-Statistics-Masters-Thesis/tables/cvae_architecture_table.tex"

with open(output_path, 'w') as f:
    f.write(latex_table)

print(f"LaTeX table saved to {output_path}")
```
