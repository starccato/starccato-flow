```python
from starccato_flow.training.trainer_cvae import ConditionalVAETrainer
from starccato_flow.data import sTheta
```

### Configs

```python
font_family = "serif"
font_name = "Times New Roman"
```

```python
data = sTheta()
```

```python
trainer = ConditionalVAETrainer(
    num_epochs=256,
    validation_split=0.1
)
trainer.load_pretrained(model_path="../outdir/cvae_weights.pt")
```

### Generated Signal Plots (changing electron fraction)

```python
trainer._plot_signal_grid(epoch=256, fname="../../2025-Statistics-Masters-Thesis/figures/signal-generation/generated_signal_grid_varying_ye_4x4.png")
```

### Latent Space

```python
trainer._plot_latent_space(epoch=256, fname="../../2025-Statistics-Masters-Thesis/figures/signal-generation/cvae_latent_space.png")
```

### Matched Generated Distribution

```python
trainer.generate_and_plot_signal_distribution(
    num_samples=10000,
    fname="../../2025-Statistics-Masters-Thesis/figures/signal-generation/matched_generated_signal_distribution.png",
    background="white",
    font_family="serif",
    font_name="Times New Roman",
    sample_from_data=True
)

data.plot_signal_distribution()
```

### Matched Generated Distribution by Beta

```python
trainer.generate_and_plot_signal_distribution(
    num_samples=10000,
    fname="../../2025-Statistics-Masters-Thesis/figures/signal-generation/matched_generated_signal_distribution.png",
    background="white",
    font_family="serif",
    font_name="Times New Roman",
    sample_from_data=True,
    beta_min=0.07,
    beta_max=0.15
)
```
