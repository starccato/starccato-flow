#### Slow rotation

```python
beta_min_slow = 0.0
beta_max_slow = 0.06

trainer.generate_and_plot_signal_distribution(
    num_samples=10000,
    fname="../../2025-Statistics-Masters-Thesis/figures/signal-generation/matched_generated_signal_distribution.pdf",
    background="white",
    font_family="serif",
    font_name="Times New Roman",
    sample_from_data=True,
    beta_min=beta_min_slow,
    beta_max=beta_max_slow
)

data.plot_signal_distribution(beta_min=beta_min_slow, beta_max=beta_max_slow)
```

#### Rapid rotation

```python
beta_min_fast = 0.06
beta_max_fast = 0.17

trainer.generate_and_plot_signal_distribution(
    num_samples=10000,
    fname="../../2025-Statistics-Masters-Thesis/figures/signal-generation/matched_generated_signal_distribution.png",
    background="white",
    font_family="serif",
    font_name="Times New Roman",
    sample_from_data=True,
    beta_min=beta_min_fast,
    beta_max=beta_max_fast
)

data.plot_signal_distribution(beta_min=beta_min_fast, beta_max=beta_max_fast)
```

#### Extreme rotation

```python
beta_min_rapid = 0.17
beta_max_rapid = 0.5

trainer.generate_and_plot_signal_distribution(
    num_samples=10000,
    fname="../../2025-Statistics-Masters-Thesis/figures/signal-generation/matched_generated_signal_distribution.pdf",
    background="white",
    font_family="serif",
    font_name="Times New Roman",
    sample_from_data=True,
    beta_min=beta_min_rapid,
    beta_max=beta_max_rapid
)

data.plot_signal_distribution(beta_min=beta_min_rapid, beta_max=beta_max_rapid)
```
