```python
from starccato_flow.training.trainer_cvae import ConditionalVAETrainer
from starccato_flow.data import sTheta
```

### Configs

```python
font_family = "serif"
font_name = "Times New Roman"
```

```python
data = sTheta()
```

```python
trainer = ConditionalVAETrainer(
    num_epochs=256,
    validation_split=0.1
)
trainer.load_pretrained(model_path="../outdir/cvae_weights.pt")
```
