# Toy Flow Matching Demo

This notebook shows a simple 2D Gaussian toy example of flow matching. The samples move from a broad Gaussian prior toward a structured posterior with multiple moon-shaped clusters over timesteps 0.0 to 1.0.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import to_rgba

from starccato_flow.utils.plotting_defaults import GENERATED_SIGNAL_COLOUR

np.random.seed(7)
```

```python
# Mock flow-matching samples for theta_0 and theta_1.
timesteps = [0.0, 0.25, 0.5, 0.75, 1.0]

# Start from a broad Gaussian prior.
prior = np.random.normal(loc=0.0, scale=1.0, size=(900, 2))

# Build a simple target posterior with multiple moon-shaped clusters.
angles = np.linspace(0, 2 * np.pi, 900)
moon_a = np.column_stack([
    np.cos(angles) * 0.9 + 0.25,
    np.sin(angles) * 0.55 - 0.15,
])
moon_b = np.column_stack([
    -np.cos(angles) * 0.9 - 0.25,
    np.sin(angles) * 0.55 + 0.15,
])

target = np.vstack([moon_a[:450], moon_b[:450]])
target = target + 0.03 * np.random.normal(size=target.shape)

# Interpolate between prior and posterior across timesteps.
posterior_samples = {}
for t in timesteps:
    if t == 0.0:
        samples = prior.copy()
    elif t == 1.0:
        samples = target.copy()
    else:
        blend = t
        bridge = (1 - blend) * prior[:target.shape[0]] + blend * target
        bridge += 0.03 * np.random.normal(size=bridge.shape)
        samples = bridge
    posterior_samples[t] = samples

true_theta = np.array([0.0, 0.0])
```

```python
def plot_flow_matching_evolution(samples_by_t, true_theta, timesteps, fname=None):
    plt.rcParams.update({
        "font.family": "Avenir",
        "axes.titlesize": 16,
        "axes.labelsize": 14,
        "xtick.labelsize": 11,
        "ytick.labelsize": 11,
        "figure.facecolor": "none",
        "axes.facecolor": "none",
        "savefig.facecolor": "none",
        "savefig.transparent": True,
    })

    fig, axes = plt.subplots(1, len(timesteps), figsize=(16, 3.6), sharex=True, sharey=True)
    fig.patch.set_alpha(0.0)

    for ax, t in zip(axes, timesteps):
        samples = samples_by_t[t]
        ax.scatter(samples[:, 0], samples[:, 1], s=12, alpha=0.45, c=GENERATED_SIGNAL_COLOUR, edgecolors="none")
        ax.set_title(f"t = {t:.2f}", color="white", pad=6)
        ax.set_xlabel(r"$\theta_0$", color="white", fontsize=15)
        ax.set_ylabel(r"$\theta_1$", color="white", fontsize=15)
        ax.grid(False)
        ax.set_facecolor("none")
        for spine in ax.spines.values():
            spine.set_color("white")

    shared_limits = (-2.2, 2.2)
    for ax in axes:
        ax.set_xlim(*shared_limits)
        ax.set_ylim(*shared_limits)
        ax.tick_params(colors="white", labelsize=11)

    plt.tight_layout()
    if fname is not None:
        plt.savefig(fname, dpi=200, bbox_inches="tight", transparent=True)
    plt.show()

plot_flow_matching_evolution(posterior_samples, true_theta, timesteps, fname="../poster-presentation/toy_flow_matching_evolution.svg")
```
